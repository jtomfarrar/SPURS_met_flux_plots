```python
import numpy as np
# from netCDF4 import Dataset
# from netCDF4 import num2date
#import netCDF4 as nc
import matplotlib.pyplot as plt
import matplotlib as mplt
# from scipy import signal
# import datetime
# import nc_time_axis
import xarray as xr
```

```python
%matplotlib qt 
plt.rcParams['figure.figsize'] = (5,4)
plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 400
plt.close('all')

__figdir__ = '../plots/' + 'SPURS1_'
savefig_args = {'bbox_inches':'tight', 'pad_inches':0.2}
```

```python
savefig = True
zoom = False
if zoom:
    tmin, tmax = (220,245)
else:
    tmin, tmax = (220,245)
```

```python
# Load SPURS1 data

# If using local path
#Define path using the r prefix (which means raw string so that special character / should not be evaluated)
# datapath=r"C:/D_drive/SPURS/spurs1_for_distribution/distributed/"
# filename = datapath+"SPURS_WHOI_1_D_M1H.nc"

# Get data from NASA JPL PO.DAAC
filename = 'https://podaac-opendap.jpl.nasa.gov/opendap/allData/insitu/L2/spurs1/mooring/SPURS_WHOI_1_D_M.nc'
```

```python
dsmerged = xr.open_dataset(filename)

timeW = dsmerged.TIME  # 'days since 1950-01-01 00:00:00'

fig = plt.figure(figsize=(6,4))

plt.subplot(4,1,1)
plt.plot(timeW,dsmerged.TEMP)
plt.ylabel('SST ($^\circ$C)')
plt.title('Met summary at ' + str(round(float(dsmerged.LATITUDE),4)) + '$^\circ$N, ' + str(round(float(dsmerged.LONGITUDE),4)) + '$^\circ$E')
var='SST'
plt.legend(['Buoy '+var,'ERA5 '+var], loc='upper right')
ax = plt.gca()
ax.set_xticklabels([])

plt.subplot(4,1,2)
plt.plot(timeW,dsmerged.AIRT)
plt.ylabel('Air temp ($^\circ$C)')
var='air temp'
# plt.legend(['Buoy '+var,'ERA5 '+var], loc='upper right')
ax = plt.gca()
ax.set_xticklabels([])

plt.subplot(4,1,3)
plt.plot(timeW,np.sqrt(dsmerged.UWND**2+dsmerged.VWND**2))
plt.ylabel('Wind speed (m/s)')
var='wind speed'
ax = plt.gca()
ax.set_xticklabels([])

plt.subplot(4,1,4)
plt.plot(timeW,dsmerged.UWND)
plt.plot(timeW,dsmerged.VWND)
plt.legend(['Buoy U', 'Buoy V','ERA5 U', 'ERA5 V'], loc='upper right')
plt.ylabel('U, V (m/s)')

if savefig:
    plt.savefig(__figdir__+'met_summary.pdf',**savefig_args)

fig = plt.figure(figsize=(6,4))
n, bins, patches = plt.hist(np.sqrt(dsmerged.UWND**2+dsmerged.VWND**2),30,density=True,edgecolor='k') # 
plt.title('Wind speed PDF')
plt.xlabel('Wind speed')
plt.ylabel('Probablility density')

if savefig:
    plt.savefig(__figdir__+'wind_PDF.pdf',**savefig_args)
```

```python
fig = plt.figure(figsize=(6,4))

plt.plot(timeW,dsmerged.SW)
plt.ylabel('Solar radiation (W/m$^2$)')
plt.title('Solar radiation at ' + str(round(float(dsmerged.LATITUDE),4)) + '$^\circ$N, ' + str(round(float(dsmerged.LONGITUDE),4)) + '$^\circ$E')
var='SW'
# plt.legend(['Buoy '+var,'ERA5 '+var], loc='upper right')
```

```python
# If desired-- use gui zoom on figure and re-run this cell
fig.autofmt_xdate()
if savefig:
    plt.savefig(__figdir__+'SWR.pdf',**savefig_args)
```
